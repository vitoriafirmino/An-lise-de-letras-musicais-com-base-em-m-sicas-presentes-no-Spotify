```python
import re
def limpar_texto(musica):
    if isinstance(musica, str):
        musica = re.sub(r'\n+', ' ', musica)  # Remove os caracteres "\n"
        musica = re.sub(r'\s+', ' ', musica).strip()  # Remove espaços extras
    return musica

dfFamosas['Letra da Música'] = dfFamosas['Letra da Música'].apply(limpar_texto)
dfFamosas.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(data=dfFamosas, x='Gênero Musical')
plt.title('Distribuição de Gêneros Musicais - Famosas')
plt.xlabel('Gênero Musical')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(15,6))
sns.countplot(data=dfFamosas, x='Ano de Lançamento')
plt.title('Distribuição Musical nos Anos')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
